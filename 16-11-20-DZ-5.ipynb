```
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def people():
    numbers = input('\nВведите номер документа:')
    for doc_numbers in documents:
        if doc_numbers["number"] == numbers:
            print('Результат:')
            print(f'Владелец документа: {doc_numbers["name"]}')
            break
    else:
        print('Результат:')
        print('Документ не найден в базе')

def shelf():
    numbers = input('\nВведите номер документа:')
    break_marker = False
    for shelf_directories in directories.items():
        for doc_numbers in shelf_directories[1]:
            if doc_numbers == numbers:
                print('Документ хранится на полке:', shelf_directories[0])
                break_marker = True
                break
        if break_marker == True:
            break
    else:
        print('Результат:')
        print('Документ не найден в базе')

def shelf_2(numbers):
    for shelf_directories in directories.items():
        for doc_numbers in shelf_directories[1]:
            if doc_numbers == numbers:
                return shelf_directories[0]

def people_list():
    print('\nРезультат:')
    print('\n')
    for persons in documents:
        print(f"№: {persons['number']}, тип: {persons['type']}, владелец: {persons['name']}, полка хранения: {shelf_2(persons['number'])}")

def as_command():
    number = input('\nВведите номер полки:')
    for shelf_directories in directories.items():
        if number in shelf_directories[0]:
            print('\nРезультат:')
            print(f'Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}')
            break
    else:
        directories[number] = []
        print('\nРезультат:')
        print(f'Полка добавлена. Текущий перечень полок: {list(directories.keys())}')

def delete_shelf():
    number = input('Введите номер полки: ')
    if number in list(directories.keys()) and directories[number] == []:
        del directories[number]
        print('\nРезультат:')
        print(f'Полка удалена. Текущий перечень полок: {list(directories.keys())}')
    elif number in directories and directories[number] != []:
        print('\nРезультат:')
        print(f'На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: {list(directories.keys())}')
    else:
        print('\nРезультат:')
        print(f'Такой полки не существует. Текущий перечень полок: {list(directories.keys())}')

def example():
    while True:
        command = input('\nВведите команду: ')
        if command == 'p':
            people()
        elif command == 's':
            shelf()
        elif command == 'l':
            people_list()
        elif command == 'as':
            as_command()
        elif command == 'ds':
            delete_shelf()
        elif command == 'q':
            break
        else:
            print('Вы ввели некорректную команду, повторите ввод.')

example()
```
